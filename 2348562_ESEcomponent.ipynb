```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Iris.csv')
```

```python
# Display the first few rows of the dataset
print(df.head())

# Get summary statistics
print(df.describe())
```

```python
df.hist(figsize=(10, 10))
```

```python
# Create a pair plot with different colors for each species
sns.set_style("whitegrid")
sns.pairplot(df, hue="Species", markers=["o", "s", "D"])
plt.show()
```

```python
features = ['PetalLengthCm', 'PetalWidthCm']
X = df[features]

# Target variable
y = df['Species']

# Split the data into training (70%) and testing (30%) sets
# Random_state is set for reproducibility, you can change it or omit it
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Printing the shapes of the resulting sets for verification
print("Training set shape (X, y):", X_train.shape, y_train.shape)
print("Testing set shape (X, y):", X_test.shape, y_test.shape)
```

```python

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score

# Create a logistic regression model
model = LogisticRegression(max_iter=1000, multi_class='ovr')

# Fit the model to the training data
model.fit(X_train, y_train)

# Predict the iris species on the testing data
y_pred = model.predict(X_test)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Precision
precision = precision_score(y_test, y_pred, average='weighted')
print(f'Precision: {precision:.2f}')
```
